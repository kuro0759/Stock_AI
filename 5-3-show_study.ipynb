```python
from matplotlib import pyplot as plt
import numpy as np
import optuna
import pickle
```

```python
# 定数
FILENAME = 'study.pkl'
```

```python
# pickleファイルからデータを復元
with open(FILENAME, 'rb') as f:
    study = pickle.load(f)
print(f'best_params = {study.best_params}')
print(f'best_value = {study.best_value}')
print(f'number of trials = {len(study.trials)}')
```

```python
# trialのデータ構造を調べる
print(f'trials[0] = {study.trials[0]}')
```

```python
# トライアル番号の一覧
numbers = [trial.number for trial in study.trials]
print(f'numbers = {numbers}')
```

```python
# value(誤差)の一覧
values = np.asarray([trial.value for trial in study.trials])
print(f'values = {values}')
```

```python
import time, optuna, plotly.io as pio

plots = [
    ("timeline"   , optuna.visualization.plot_timeline),
    ("history"    , optuna.visualization.plot_optimization_history),
    ("importances", optuna.visualization.plot_param_importances),
    ("slice"      , optuna.visualization.plot_slice),
]

for name, fn in plots:
    t0  = time.perf_counter()
    fig = fn(study)
    fig.write_html(f"{name}.html", include_plotlyjs="cdn")   # ★HTMLのみ出力
    fig.show()                                               # Notebook で確認
    print(f"{name:11s}: {(time.perf_counter()-t0):.2f} s")
```
